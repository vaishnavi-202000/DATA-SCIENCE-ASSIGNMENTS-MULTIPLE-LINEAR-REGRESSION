```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot

df = pd.read_csv("50_Startups.csv")
df.shape
```

```python
# Model Building
```

```python
model=smf.ols("Profit~RDS+ADMS+MKTS",data=df1).fit()
```

```python
# Model Testing
# Finding Coefficient parameters
model.params
```

```python
# Finding tvalues and pvalues
model.tvalues , np.round(model.pvalues,5)
```

```python
model.rsquared , model.rsquared_adj
```

```python
# Build SLR(Simple Linear Regression) and MLR(Multiple Linear Regression) models for insignificant variables 'ADMS' and 'MKTS'
# Also find their tvalues, pvalues and rsquared values.
```

```python
slr_a=smf.ols("Profit~ADMS",data=df1).fit()
slr_a.tvalues , slr_a.pvalues
```

```python
slr_a.rsquared , slr_a.rsquared_adj
```

```python
slr_m=smf.ols("Profit~MKTS",data=df1).fit()
slr_m.tvalues , slr_m.pvalues
```

```python
slr_m.rsquared , slr_m.rsquared_adj
```

```python
mlr_am=smf.ols("Profit~ADMS+MKTS",data=df1).fit()
mlr_am.tvalues , mlr_am.pvalues
```

```python
mlr_am.rsquared , mlr_am.rsquared_adj
```

```python
#Residual Analysis
# Test for Normality of Residuals (Q-Q Plot) using residual model (model.resid)
```

```python
sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
model.resid.hist()
```

```python
# Test for Homoscedasticity or Heteroscedasticity (plotting model's standardized fitted values vs standardized residual values)
```

```python
def standard_values(vals) : return (vals-vals.mean())/vals.std()

plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('Fitted values')
plt.ylabel('residual values')
plt.show()
```

```python
# Model Deletion Diagnostics
## Detecting Influencers/Outliers
## Cook’s Distance

model_influence = model.get_influence()
model_influence
(cooks, pvalue) = model_influence.cooks_distance
cooks = pd.DataFrame(cooks)
cooks[0].describe()
```

```python
#Plot the influencers values using stem plot
import matplotlib.pyplot as plt
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(df1)), np.round(cooks[0], 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
np.argmax(cooks) , np.max(cooks)
```

```python
## High Influence points
## Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers
influence_plot(model)
plt.show()
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k=df1.shape[1]
n=df1.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

```python
df1[df1.index.isin([49])]
```

```python
# Improving the Model
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
df2=df1.drop(df1.index[[49]],axis=0).reset_index(drop=True)
df2
```

```python
# Model Deletion Diagnostics and Final Model

while np.max(cooks)>0.5 :
    model=smf.ols("Profit~RDS+ADMS+MKTS",data=df2).fit()
    (cooks,_)=model.get_influence().cooks_distance
    cooks
    np.argmax(cooks) , np.max(cooks)
    df2=df2.drop(df2.index[[np.argmax(cooks)]],axis=0).reset_index(drop=True)
    df2
else:
    final_model=smf.ols("Profit~RDS+ADMS+MKTS",data=df2).fit()
    final_model.rsquared , final_model.aic
    print("Thus model accuracy is improved to",final_model.rsquared)

final_model.rsquared

df2
```
